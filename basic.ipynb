```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([10, 20, 30, 40, 50])

# Manipulating arrays (reshaping)
matrix = np.arange(1, 10).reshape(3, 3)

# Basic arithmetic operations
sum_arrays = arr1 + arr2  # Element-wise addition
diff_arrays = arr1 - arr2  # Element-wise subtraction
prod_arrays = arr1 * arr2  # Element-wise multiplication
div_arrays = arr1 / arr2   # Element-wise division

# Aggregations
mean_value = np.mean(arr1)     # Mean of arr1
median_value = np.median(arr1) # Median of arr1
sum_value = np.sum(arr1)       # Sum of arr1

# Display results
print("Array 1:", arr1)
print("Array 2:", arr2)
print("Sum of Arrays:", sum_arrays)
print("Difference of Arrays:", diff_arrays)
print("Product of Arrays:", prod_arrays)
print("Division of Arrays:", div_arrays)
print("Matrix:\n", matrix)
print("Mean:", mean_value)
print("Median:", median_value)
print("Sum:", sum_value)
```

```python
grouped_data = df.groupby("Department")["Salary"].mean()

# Bar plot for aggregated data
grouped_data.plot(kind="bar", color='c', figsize=(6, 4))
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.xticks(rotation=45)
plt.show()

# Data Cleaning: Introduce and handle missing values
df.loc[2, "Salary"] = np.nan
df["Salary"].fillna(df["Salary"].median(), inplace=True)

# Combined Subplots (Line, Bar, Scatter)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Line Plot
axes[0].plot(df["Age"], df["Salary"], marker='o', color='b')
axes[0].set_title("Age vs Salary (Line Plot)")

# Bar Plot
axes[1].bar(df["Name"], df["Salary"], color='g')
axes[1].set_title("Salary by Employee")
axes[1].tick_params(axis='x', rotation=45)

# Scatter Plot
axes[2].scatter(df["Age"], df["Salary"], color='r')
axes[2].set_title("Age vs Salary (Scatter Plot)")

plt.tight_layout()
plt.show()
```

```python
# Custom Visualization: Styled Line Plot
plt.figure(figsize=(8, 5))
plt.plot(df["Age"], df["Salary"], linestyle="--", marker="s", color="purple", linewidth=2, markersize=8, label="Styled Line")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Customized Plot of Age vs Salary")
plt.legend()
plt.grid(True)
plt.show()

# EDA: Box Plot for Salary Distribution
plt.figure(figsize=(6, 4))
sns.boxplot(x=df["Salary"])
plt.title("Salary Distribution")
plt.show()

# Histogram with Density Plot
plt.figure(figsize=(6, 4))
sns.histplot(df["Age"], kde=True, bins=10)
plt.title("Age Histogram with Density Plot")
plt.show()
```
